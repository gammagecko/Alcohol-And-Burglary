```python
# Data Source: https://www.gigasheet.com/sample-data/per-capita-us-alcohol-consumption
filename = r"C:\Users\PC\Documents\Crime Data\state_alcohol.csv"
df2 = pd.read_csv(filename)
df2.head()
```

```python
df2 = df2.sort_values(by=['State', 'Year'])
keep = ['Year', 'State', 'Beverage_Type', 'Gallons_of_Beverage']
df2 = df2[keep]
df2.head()
```

```python
df2 = df2.groupby(['Year', 'State'])['Gallons_of_Beverage'].sum().reset_index()
df2.head()
```

```python
df2.rename(columns={'Gallons_of_Beverage': 'Gallons_of_Alcoholic_Beverage'}, inplace=True)
mask = df['Year'].between(1970, 2019, inclusive='both')
df = df[mask]
mask = df2['Year'].between(1970, 2019, inclusive='both')
df2 = df2[mask]
print(df.shape[0])
print(df2.shape[0])
```

```python
df3 = pd.merge(df, df2, on=['Year', 'State'], how='inner')
print(df3)
```

```python
df5['Year'] = pd.to_numeric(df5['Year'], errors='coerce')
df3 = pd.merge(df3, df5, on=['Year'], how='left')
print(df3)
```

```python
df3['Alcohol Per Capita'] = df3['Gallons_of_Alcoholic_Beverage'] / df3['Population']
print(df3)
```

```python
df3['prev_burg'] = df3['Burglary Rate'].shift(1)
df3['Burglary 1 Yr Growth %'] = df3['Burglary Rate'] / df3['prev_burg']
df3['prev_alcohol'] = df3['Alcohol Per Capita'].shift(1)
df3['Alcohol Per Capita 1 Yr Growth %'] = df3['Alcohol Per Capita'] / df3['prev_alcohol']
df3['prev_population'] = df3['Population'].shift(1)
df3['Population 1 Yr Growth %'] = df3['Population'] / df3['prev_population']
df3.drop(columns=['prev_burg', 'prev_alcohol', 'prev_population'], inplace=True)
```
